```python
# open all page  

def open_nextpage():  
    pagetocsv()
    try:
        postpages=driver.find_elements_by_xpath(".//*[@class='ipsField_fullWidth' and @type='number']")[0]
        postpages=postpages.get_attribute('max')
        try:
#             print(len(postpages))
            print(postpages.get_attribute('max'))
            
            print('chau')
        except:
            postpages=driver.find_elements_by_xpath(".//*[@class='ipsBox ipsResponsive_pull ipsPadding:half ipsClearfix ipsClear ipsMargin_bottom']/ul")[0].get_attribute('data-pages')
    #         postpages
            print('hola')
    #         print(len(postpages))
        # print(postpages.get_attribute('max'))
        actpostpage=driver.find_elements_by_xpath(".//*[@class='ipsPagination_page ipsPagination_active']/a")[0].text

        print('mas de una pagina   ' + 'pagina actual:  '+ str(actpostpage))
        
    except:
        postpages='1'
        actpostpage='1'
        print(str(actpostpage) + " of " +str(postpages))
    
    next_page = driver.find_elements_by_xpath(".//*[@rel='next']")
    # 
    
    try:
        print(next_page[0].get_attribute('href') + '   proxima pagina')
        next_page_=next_page[0].get_attribute('href')
        print('mas de una pagina shelav 2')
        print(postpages)

        handles = len(driver.window_handles)

        sleep(3)

        try:
            if actpostpage!=postpages:
                print('not last page')
                driver.execute_script('''window.open("","_blank");''')
    #             Print(len(driver.window_handles))
                sleep(1)
                driver.switch_to.window(driver.window_handles[(len(driver.window_handles)-1)])
#                 print(next_page_)
                driver.get(next_page_)
                sleep(3)
                print(str(actpostpage) + " of " +str(postpages))
            else:
                print('termino el loop de paginas')
                print(str(actpostpage) + " of " +str(postpages))
                global finished
                finished=True
        except:
            print('fallo apertura next_page')
    except:
        finished=True
        print("una sola pagina existe")
        print(str(actpostpage) + " of " +str(postpages))
```
